# Checking the distribution of class

```python
classes = df['Class'].value_counts()
classes
```

```python
normal_share = classes[0] / df['Class'].count()
normal_share
```

```python
fraud_share = classes[1] / df['Class'].count()
fraud_share
```

If we convert the normal_share and fraud_share into percentages then the normal_share is 99.82% and the fraud_share is only 0.17%.

```python
# Bar plotting the number of fraudulent vs non-fraudulent transcations
sns.countplot(x='Class',data = df)
plt.title('No of Non-fraudulent vs fraudulent transcations')
plt.show()
```

```python
#bar plotting the percentage of fraudulent and non-fraudulent transactions
fraud_percentage = {'Class':['Non-Fraudulent', 'Fraudulent'], 'Percentage':[normal_share, fraud_share]} 
df_fraud_percentage = pd.DataFrame(fraud_percentage) 
sns.barplot(x='Class',y='Percentage', data=df_fraud_percentage)
plt.title('Percentage of Non-fraudulent vs fraudulent transcations')
plt.show()
```

# Observe the distribution of classes

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating non fradulent dataframe
data_non_fraud = df[df['Class'] == 0]
```

```python
#Creating fraudulent dataframe
data_fraud = df[df['Class'] == 1]
```

```python
#Plotting distribution
plt.figure(figsize = (8,5))
a = sns.distplot(data_non_fraud['Time'],label = 'Non fraudulent',hist=False)
a = sns.distplot(data_fraud['Time'],label = 'fraudulent',hist = False)
a.set(xlabel = 'Time B/W the transaction and the first transaction')
plt.legend()
plt.show()
```

We do not see any pattern for the non-fraudulent and fraudulent transactions with respect to time.So,we can drop the Time column

```python
# Dropping the Time column
df.drop('Time',axis=1,inplace = True)
```

# Splitting the Data set 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Class'],axis=1)
```

# Model Creation

# By using logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
l = LogisticRegression()
```

```python
model = l.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

# Model's performance considering Oversampling

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
```

```python
smote = SMOTE(random_state = 0)
```

```python
X1_resampled, y1_resampled = smote.fit_resample(X1_train, y1_train)
```

```python
model2 = RandomForestClassifier()
model2.fit(X1_resampled, y1_resampled)
```

```python
y_pred_resampled = model2.predict(X1_test)
```

```python
print(confusion_matrix(y1_test,y_pred_resampled))
```

```python
print(classification_report(y1_test,y_pred_resampled))
```
